```python
## load libraries
import pandas as pd
import numpy as np
```

```python
## load data
## we will be using the titanic dataset

# Create URL
### url = 'https://raw.githubusercontent.com/chrisalbon/simulated_datasets/master/titanic.csv'


## load the data
### data_frame = pd.read_csv(url)

### Write the data to the local disk


## data_frame.to_csv('dataframe.csv')

## load the data from the local disk
dataframe = pd.read_csv('dataframe.csv', index_col="Unnamed: 0")
dataframe.head()
```

```python
## dataset properties
dataframe.info()
```

>- The dataset has 6 variables and 1313 entries

### Row Selection based on Conditions

```python
## Select the female passengers
dataframe[dataframe['Sex'] == 'female']
```

```python
## filter with more than two conditions
dataframe[(dataframe['Sex'] == 'female') & (dataframe['PClass'] == '1st')] 
```

```python
## filter using numeric variables
## For age greater that 55 years
dataframe[dataframe['Age'] > 55].head(5)
```

### Value Replacement

```python
## Replace male with man
dataframe.replace('male', 'man').head()
```

```python
## Replace more than one value of the same variable
dataframe.replace(['male', 'female'], ['man', 'woman'])
```

```python
## Renaming columns
dataframe.rename(columns={'Sex': 'Gender'})
```

```python
## Rename multiple ; do this by supplying a dictionary of the new-name:old-name
dataframe.rename(columns={'Sex': 'Gender', 'PClass':'Passenger_Class'})
```

```python
## Total missing observations by columns
dataframe.isnull().sum()
```

### Categorical Variables

```python
## Unique values
dataframe['Sex'].unique()
```

```python
## Value counts for the categories
dataframe['Sex'].value_counts()
```

```python
## Value counts for the categories(Proportion)
dataframe['Sex'].value_counts(normalize=True)
```

### Handling Missing Values

```python
## Check for the missing values for each column in the data frame
dataframe.isnull().sum()
```

>- There is only one variable that contains missing observations(557)

```python
## replace missing observations with -999
dataframe.replace(np.NaN, -999)
```

### Column deletion

```python
## delete a single column by it's name
dataframe.drop('SexCode', axis=1)
```

```python
## drop multiple columns
dataframe.drop(['Sex', 'Name', 'SexCode'], axis=1)
```

### Deleting (Duplicate) Rows

>- Deleting rows is similar to filtering out rows

```python
## Filter out the 2nd Class Passengers
dataframe[dataframe['PClass'] != '2nd']
```

```python
## Dropping duplicate rows
dataframe.drop_duplicates()
```

### Grouping Rows by Values

```python
## Group by Passenger Class
dataframe.groupby('PClass').sum()
```

```python
## Restrict the operation to a specific columns
dataframe[['PClass', 'Age']].groupby('PClass').mean('Age')
```

```python
## The Above can also be performed alternatively as below:
dataframe.groupby('PClass')['Age'].mean()
```

```python
## Grouping by multiple Columns
dataframe.groupby(['PClass', 'Sex'])[['Age', 'SexCode']].sum()
```

```python
## drop the * row from the data frame
df = dataframe[dataframe['PClass'] != '*']
```

```python
## Value counts for the df[PClass]
df['PClass'].value_counts()
```

### Looping over Columns

```python
## List comprehensions
## return the first five entries after using the list comprehensions
[name.upper() for name in df['Name']][0:4]
```

### Apply function

```python
df['Name'].apply(str.upper)
```

```python
## Arranging data sorting
print('Option1\n', df.value_counts(['PClass', 'Survived']))

## The above can also be alternatively be performed as follows
print('Option2 for the above\n', df[['PClass', 'Survived']].value_counts())

## You can also implement hierarchical indexing as follows
print('Display results by descending order\n', df[['PClass', 'Survived']].value_counts().sort_index(ascending = False))

## sort the results
print('return the default results\n', df['PClass'].value_counts())

## By default the above sorts the results in descending order
## Removing the default sort
print('Don\'t Sort by default\n', df['PClass'].value_counts(sort= False))
```

```python
## select the numeric variables
df_cont = df.select_dtypes('float')
df_cont
```
